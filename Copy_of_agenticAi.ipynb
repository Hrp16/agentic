```python
from pydantic import BaseModel, Field
from langchain_core.output_parsers import PydanticOutputParser # Corrected import path

class Developer(BaseModel):
    name: str = Field(description="Developer name")
    experience: int = Field(description="Years of experience")
    skills: list[str] = Field(description="Skills list")

parser = PydanticOutputParser(pydantic_object=Developer)
```

```python
format_instructions = parser.get_format_instructions()
```

```python
prompt = f"""
Extract information from the text.

{format_instructions}

Text:
Harsh has 5 years of experience in React and TypeScript.
"""
```

```python
from huggingface_hub import InferenceClient

client = InferenceClient(api_key="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")

response = client.chat.completions.create(
    model="Qwen/Qwen2.5-7B-Instruct",
    messages=[
        {"role": "user", "content": prompt}
    ]
)

output = response.choices[0].message.content
print(output)
```
